# **Dataset: Empathetic Dialogues Dataset**

### Dataset: Empathetic Dialogues

**Source**: [Empathetic Dialogues on HuggingFace](https://huggingface.co/datasets/empathetic_dialogues) or [Original Paper](https://arxiv.org/abs/1811.00207).

**Description**:  
This dataset contains around 25,000 conversations where each conversation revolves around someone describing an emotional event and another person responding empathetically. Essentially, it models human interaction around emotions, making it ideal for creating chatbots that detect and respond empathetically based on a given situational context.

**Features**:
- **Context**: The prompt that sets the emotional event described by the user.
- **Utterance**: The user's statement in the conversation.
- **Speaker**: Labels identifying who speaks (e.g., Speaker 1, Speaker 2).
- **Emotion**: The primary emotion in the dialog (e.g., joy, sadness, anger).
- **Reactions**: The empathetic response made by the other speaker.

## **Fine-Tuning Techniques:**

### Fine-tuning Techniques: Prompt Tuning

We will now focus on **Prompt Tuning**, which is effective techniques for fine-tuning large language models (LLMs) for specific tasks, such as building empathetic therapy chatbots.

---

#### (i) Prompt Tuning:

**Prompt Tuning** involves adjusting the prompts and trainable input embeddings during fine-tuning. In this method, you create special "trainable" prompt tokens that are prepended to the input to help guide the LLM toward the desired behavior. These "trainable" prompts allow the model to emphasize or focus on certain aspects of input data during its prediction process.

Trainable prompts can be constructed by transforming existing dataset pairs into carefully designed prompts, which may look like:

- **Input**: Context or emotional event as described by the user.
- **Output**: Empathetic responses generated from trained prompt embeddings.

This method enables the LLM to respond in a predictable and task-oriented manner by modulating the behavior through these prompt adjustments during training.

---

```python
import torch
from torch.utils.data import DataLoader
from datasets import Dataset
import pandas as pd
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments

# Load the tokenizer for the lightweight DistilGPT-2 model
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")

# Set the pad_token to eos_token (or add a custom pad_token if preferred)
tokenizer.pad_token = tokenizer.eos_token

# Tokenize the input-output pairs for training and testing data
train_encodings = tokenizer(train_df['input'].tolist(), truncation=True, padding=True, max_length=128)
test_encodings = tokenizer(test_df['input'].tolist(), truncation=True, padding=True, max_length=128)

# Create a HuggingFace Dataset from tokenized encodings using both input_ids and attention_mask
train_dataset = Dataset.from_dict({
    'input_ids': train_encodings['input_ids'],
    'attention_mask': train_encodings['attention_mask']
})

test_dataset = Dataset.from_dict({
    'input_ids': test_encodings['input_ids'],
    'attention_mask': test_encodings['attention_mask']
})

# Load the pre-trained light model (DistilGPT-2) for fine-tuning
model = AutoModelForCausalLM.from_pretrained("distilgpt2")

# Move model to GPU if available
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)

# Prepare DataLoader for fine-tuning (with tokenized inputs in batches of 8)
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8)
```

```python
# Define a function to preprocess the dataset and add 'labels'
def preprocess_data_for_clm(examples):
    inputs = examples['input_ids']
    # For causal language modeling, the labels are just the inputs, but it will shift during training
    outputs = examples['input_ids']

    examples['labels'] = outputs  # Set labels to be the same as input_ids
    return examples

# Apply preprocessing to the dataset (train and test): Add 'labels'
train_dataset = train_dataset.map(preprocess_data_for_clm, batched=True)
test_dataset = test_dataset.map(preprocess_data_for_clm, batched=True)
```

```python
from transformers import TrainingArguments

# Update training arguments to include more frequent logging steps
training_args = TrainingArguments(
    output_dir="./results",
    overwrite_output_dir=True,
    num_train_epochs=3,  # Using 3 epoch for quicker training
    per_device_train_batch_size=8,
    save_steps=100,
    logging_steps=10,  # Log more frequently
    save_total_limit=2,
    logging_dir='./logs',
    report_to="none"  # To avoid connecting to external logging services like WandB
)

# Reinitialize Trainer with updated arguments
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

# Retrain or continue training with save checkpoint
trainer.train()
```

### **Part 2: Define the Fine-Tuning Configuration**

To fine-tune, we must configure key parameters such as the optimizer, learning rate, and loss function, and then iterate through each batch to perform training.

```python
from transformers import DataCollatorForLanguageModeling

# Create a data collator for language modeling tasks like GPT
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # False for causal language modeling (CLM), True for masked LM (BERT)
)

# Re-create the DataLoader with the collator that returns tensors
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=data_collator)
test_loader = DataLoader(test_dataset, batch_size=8, collate_fn=data_collator)
```

```python
# Custom collate function to handle plain lists and convert to tensors
def collate_fn(batch):
    input_ids = torch.tensor([example['input_ids'] for example in batch], dtype=torch.long)
    attention_mask = torch.tensor([example['attention_mask'] for example in batch], dtype=torch.long)
    return {"input_ids": input_ids, "attention_mask": attention_mask}

# Re-create the DataLoader with the custom collator
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=8, collate_fn=collate_fn)
```

```python
import torch
from torch.optim import AdamW

# Create an optimizer
optimizer = AdamW(model.parameters(), lr=2e-5)

# Move model to the device
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)

# Number of epochs
epochs = 3

# Training loop
for epoch in range(epochs):
    model.train()

    # Iterate over batches in the DataLoader
    for step, batch in enumerate(train_loader):
        optimizer.zero_grad()

        # Move input tensors to the device (input_ids and attention_mask)
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = input_ids  # Use inputs as labels for causal LM

        # Forward pass
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)

        # Loss computation
        loss = outputs.loss

        # Backpropagation
        loss.backward()

        # Optimization step
        optimizer.step()

        # Print loss every 100 steps
        if step % 100 == 0:
            print(f"Epoch: {epoch} | Step: {step} | Loss: {loss.item()}")

# Save the fine-tuned model
model.save_pretrained("./empathetic_chatbot_finetuned")
```

# Retrieval-Augmented Generation (RAG) with GPT-2

### Concept Overview:

**Retrieval-Augmented Generation (RAG)** combines two powerful techniques:
1. **Retrieval-based systems**: Retrieve relevant past information based on a query.
2. **Generative models**: Use models like GPT-2 to generate responses.

In RAG, we first retrieve relevant past dialogues, then use this information to help a generative model like GPT-2 create better, more contextually relevant responses.

---

### Pipeline Steps:

1. **Dataset Preparation**:
   - We load a subset of the **Empathetic Dialogues** dataset.
   - Combine the emotional prompt (context) and the response (utterance) into a single text input.

2. **Embedding Creation**:
   - Use **SentenceTransformer** to convert dialogue texts into vector embeddings.
   - These embeddings represent the meaning of the dialogues in a high-dimensional space, allowing for efficient similarity search.

3. **Retrieval System**:
   - We use **FAISS** (Facebook AI Similarity Search) to find the most similar past dialogues based on the query input.
   - FAISS performs a fast nearest neighbor search over the embeddings to retrieve the most relevant dialogues.

4. **GPT-2 Model**:
   - Fine-tune **GPT-2** to generate responses.
   - Augment the model’s input with retrieved dialogues, helping GPT-2 generate more contextually relevant responses.

5. **Prediction (Testing)**:
   - Test the system by providing a sample query (context + prompt) and generate a response using the retrieval-augmented GPT-2 model.

---

This method enhances the GPT-2 model’s ability to produce responses by leveraging relevant past dialogues, improving the relevance and quality of the generated text.


```python
import os
import random
import torch
import numpy as np
import pandas as pd
import gym
from sentence_transformers import SentenceTransformer, util
from transformers import RagTokenizer, RagRetriever, RagSequenceForGeneration
from datasets import load_dataset
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
import os
import faiss
import torch
import pandas as pd
import numpy as np
from datasets import load_dataset
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from sentence_transformers import SentenceTransformer, util

device = torch.device("cpu")
# Use GPU if available
#device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === 1. Load the empathetic dialogues dataset ===
dataset = load_dataset("empathetic_dialogues")
train_data = pd.DataFrame.from_dict(dataset['train'])

# Sample a smaller subset of the dataset
subset_size = 0.01  # Use 1% of the dataset
train_data = train_data.sample(frac=subset_size, random_state=42)

# Combine prompt (context) and utterance to create RAG-style knowledge base.
train_data['full_input'] = train_data['prompt'] + " " + train_data['utterance']
train_data = train_data[['conv_id', 'context', 'prompt', 'utterance', 'full_input']]
```

### **Create Embeddings for the Database Using Sentence-Transformers :**

```python
# === 2. Use Sentence-Transformers to Create Embeddings for Retrieval ===
# Load pre-trained SentenceTransformer model
embedder = SentenceTransformer("all-MiniLM-L6-v2", device=device)

# Embed the 'full_input' fields to use them as a retrieval source (knowledge base).
full_inputs = train_data['full_input'].tolist()
full_input_embeddings = embedder.encode(full_inputs, convert_to_tensor=True)
```

### **Build the Retrieval Index Using FAISS :**

```python
# === 3. Build the Retrieval Index (using FAISS) ===
# Initialize FAISS index for similarity lookup
d = full_input_embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(d)
faiss_index.add(full_input_embeddings.cpu().numpy())  # Add embeddings to the FAISS index
```

### **Load GPT-2 Pre-trained Tokenizer and Model :**

```python
# === 4. GPT-2 Tokenizer and Model ===
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2").to(device)
tokenizer.pad_token = tokenizer.eos_token  # Ensure proper padding
```

```python
# === 5. Retrieval + Augmentation + GPT-2 Generation ===
def retrieve_and_generate(context, prompt, emotion):
    """
    Retriever: Retrieve past utterances with similar context and generate a response.
    """
    # Join context and prompt
    input_query = prompt + " " + context

    # Embed the input query (prompt + context) for retrieval.
    input_embedding = embedder.encode(input_query, convert_to_tensor=True)

    # Search for the top K nearest neighbors.
    k = 5  # Let’s retrieve 5 most similar dialogues for augmentation.
    input_embedding_np = input_embedding.cpu().numpy()
    distances, indices = faiss_index.search(input_embedding_np[np.newaxis, :], k)

    # Collect the nearest dialogues (strings).
    retrieved_dialogues = [train_data.iloc[idx]['full_input'] for idx in indices[0]]

    # Use the retrieved dialogues as context (+ input query) for GPT-2 to generate response.
    rag_input = "\n".join(retrieved_dialogues) + f"\nQuery: {input_query}\nResponse:"

    # Tokenize the GPT-2 input and move to device (GPU/CPU).
    input_ids = tokenizer(rag_input, return_tensors="pt", max_length=512, truncation=True, padding="max_length").input_ids.to(device)

    # Generate a response using GPT-2, augmenting with the retrieved dialogues.
    outputs = model.generate(
        input_ids,
        max_new_tokens=50,  # Limit length of the response.
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode the generated result to get the final GPT-2 response.
    generated_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_response
```

```python
# === 6. Testing the Retrieval-Augmented Generation System ===
# Example input (prompt + context):
context = "sad"
prompt = "My dog died today. I am so distraught."
emotion = "sad"

# Generate response using RAG + GPT-2
response = retrieve_and_generate(context, prompt, emotion)
print(f"Generated Response:\n{response}")
```

# **Test Both RAG Model and Baseline Model**

```python
# === Baseline GPT-2 Response (No Retrieval) ===
def generate_baseline_response(context, prompt):
    """
    Generate text without retrieval, using GPT-2 directly on the input query.
    """
    # Prepare the direct input query
    input_query = prompt + " " + context

    # Tokenize the direct input query for GPT-2
    input_ids = tokenizer(input_query, return_tensors="pt", max_length=512, truncation=True, padding="max_length").input_ids.to(device)

    # Generate a response using GPT-2 without any retrieval augmentation
    outputs = model.generate(
        input_ids,
        max_new_tokens=50,  # Limit the length of the response
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode the generated result
    generated_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_response

# === Compare Baseline GPT-2 vs Retrieval-Augmented GPT-2 ===
# Test input (context + prompt)
context = "sad"
prompt = "My dog died today. I am so distraught."
emotion = "sad"  # Not used in this implementation, but can be leveraged in future extensions

# 1. RAG (Retrieval-Augmented) Response
rag_response = retrieve_and_generate(context, prompt, emotion)
print(f"--- Retrieval-Augmented GPT-2 Response: ---\n{rag_response}\n")

# 2. Baseline GPT-2 Response (No Retrieval)
baseline_response = generate_baseline_response(context, prompt)
print(f"--- Baseline GPT-2 Response: ---\n{baseline_response}\n")
```

```python
# === Present the Results in a Clear and Concise Manner ===

from tabulate import tabulate  # Install using pip if necessary: pip install tabulate

def compare_responses(context, prompt, emotion):
    """
    A function that generates and compares responses from RAG and baseline GPT-2 and prints them side by side.
    """
    # 1. Generate responses from both models
    rag_response = retrieve_and_generate(context, prompt, emotion)
    baseline_response = generate_baseline_response(context, prompt)

    # 2. Present responses side by side
    headers = ["Model", "Generated Response"]

    data = [
        ["Retrieval-Augmented GPT-2 (RAG)", rag_response],
        ["Baseline GPT-2 (No Retrieval)", baseline_response]
    ]

    print("\n===== Comparison Results =====\n")
    print(f"Context: {context}")
    print(f"Prompt : {prompt}")
    print(f"Emotion: {emotion}\n")

    print(tabulate(data, headers, tablefmt="pretty"))

# Test Example
context = "sad"
prompt = "My dog died today. I am so distraught."
emotion = "sad"

compare_responses(context, prompt, emotion)
```
